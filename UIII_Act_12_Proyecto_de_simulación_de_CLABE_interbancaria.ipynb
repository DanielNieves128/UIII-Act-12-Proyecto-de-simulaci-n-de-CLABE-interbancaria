```python
# Simulador de IA Bancaria - CLABE Interbancaria
# Programación Orientada a Objetos

class ClabeInterbancaria:

    def __init__(self, identificacion, monto, fecha_hora, banco_emisor, banco_receptor):
        self.identificacion = identificacion
        self.monto = monto
        self.fecha_hora = fecha_hora
        self.banco_emisor = banco_emisor
        self.banco_receptor = banco_receptor

    def verificar_datos_bancarios(self):
        print("\n IA: Verificando datos bancarios...")
        print("Datos verificados correctamente.")

    def transferir_fondos(self):
        print(f"Transferencia de ${self.monto:.2f} realizada.")

    def registrar_operacion(self):
        print("Operación registrada en la base de datos.")

    def notificar_transaccion(self):
        print("Notificación enviada a los bancos involucrados.")

    def confirmar_recepcion(self):
        print("✔ Fondos recibidos por el banco receptor.")


# Lista para almacenar operaciones
operaciones = []

while True:

    print("\n" + "="*50)
    print("      SISTEMA IA CLABE INTERBANCARIA")
    print("="*50)
    print("1. Registrar transferencia")
    print("2. Mostrar transferencias")
    print("3. Salir")
    print("="*50)

    opcion = input("Seleccione una opción: ")

    if opcion == "1":

        print("\n--- NUEVA TRANSFERENCIA ---")

        identificacion = input("Número de identificación: ")
        monto = float(input("Monto de transferencia: "))
        fecha_hora = input("Fecha y hora de envío: ")
        banco_emisor = input("Banco emisor: ")
        banco_receptor = input("Banco receptor: ")

        transferencia = ClabeInterbancaria(
            identificacion,
            monto,
            fecha_hora,
            banco_emisor,
            banco_receptor
        )

        print("\n--- PROCESO INTELIGENTE ---")
        transferencia.verificar_datos_bancarios()
        transferencia.transferir_fondos()
        transferencia.registrar_operacion()
        transferencia.notificar_transaccion()
        transferencia.confirmar_recepcion()

        operaciones.append(transferencia)

        print("\nTransferencia almacenada correctamente.")

    elif opcion == "2":

        if len(operaciones) == 0:
            print("\nNo existen transferencias registradas.")
        else:

            print("\nLISTADO DE TRANSFERENCIAS")
            print("-"*120)

            print(
                f"{'ID':<15}"
                f"{'MONTO':<15}"
                f"{'FECHA Y HORA':<25}"
                f"{'BANCO EMISOR':<30}"
                f"{'BANCO RECEPTOR':<30}"
            )

            print("-"*120)

            for op in operaciones:
                print(
                    f"{op.identificacion:<15}"
                    f"${op.monto:<14.2f}"
                    f"{op.fecha_hora:<25}"
                    f"{op.banco_emisor:<30}"
                    f"{op.banco_receptor:<30}"
                )

            print("-"*120)

    elif opcion == "3":
        print("\n Gracias por utilizar el Sistema IA CLABE Interbancaria.")
        break

    else:
        print("\n Opción no válida.")
```
